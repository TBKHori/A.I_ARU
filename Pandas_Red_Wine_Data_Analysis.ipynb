```python
 import pandas as pd
#Load the data directly from the internet.
wine_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
```

```python
# Print the first few lines of the dataframe
wine_df.head()
```

Selecting rows using .iloc and loc

```python
# Find the number of unique entries across each feature.
wine_df.nunique()
```

```python
# Drop some values based on the unique density feature.
wine_df = wine_df.drop_duplicates(subset='density')
```

```python
# How many entries exists?
wine_df.shape #436 unique values in the density column
```

```python
#Since density has only unique values, it can be used as index.
wine_df.set_index('density',inplace=True)
```

```python
#Check if the index changed!
wine_df.head()
```

```python
#Access the rows using iloc function.
wine_df.iloc[2]
```

```python
# Select the particular wine with density 0.9968!
wine_df.loc[0.9968]
```

```python
#Select only a few selected rows of data
wine_df.iloc[[1, 4, 7]]
```

```python
# If you find the index names, iloc may not be needed.
rows = [0.9968, 0.9964, 0.9943]
wine_df.loc[rows]
```

```python
# iloc extract a slice of data
wine_df.iloc[1:4]
```

Slice notation also works with the .loc indexer and is inclusive of the last label:

```python
# Could use variables instead.
first = 0.9970
last = 0.9959
wine_df.loc[first:last]
```

Selecting rows and columns using get_loc and index methods

```python
# Find the start and stop in that counts.
col_start = wine_df.columns.get_loc('volatile_acidity')
col_end = wine_df.columns.get_loc('volatile_acidity')+2
col_start, col_end
```

```python
# Use these new start and end as indices
wine_df.iloc[:4, col_start:col_end]
```

```python
# select the wines with the range of indicies
row_start = wine_df.index[4]
row_end = wine_df.index[7]
wine_df.loc[row_start:row_end, 'volatile_acidity': 'chlorides']
```

```python
# Find a specific combination
density_val = 0.9980
wine_df.loc[density_val, 'citric_acid']
```
