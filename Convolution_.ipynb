```python
import numpy as np
from scipy import ndimage
import cv2
from skimage import color
import matplotlib.pyplot as plt
```

```python
img = color.rgb2gray(cv2.imread('city.jpg'))
```

```python
plt.imshow(img,cmap='gray')
```

```python
img.shape
```

```python
Kx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], np.float32)
Ky = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], np.float32)
```

```python
Kx
```

```python
Ky
```

```python
filteredx = ndimage.convolve(img, Kx )
filteredy = ndimage.convolve(img, Ky )

# quivalent to sqrt(x1**2 + x2**2)   hypotenuse
filtered = np.hypot(filteredx, filteredy)
```

```python
filtered = filtered / filtered.max() * 255
filtered.shape
```

```python
type(filtered)
```

```python
filtered.max()
```

```python
filtered.min()
```

```python
plt.imshow(filtered,cmap='gray')
```

```python
# pooling process
import skimage.measure
pooled_filtered = skimage.measure.block_reduce(filtered, (2,2), np.max)
```

```python
plt.imshow(pooled_filtered,cmap='gray')
```
