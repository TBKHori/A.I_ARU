Loops or Iterations

for loop
For loop is when there is a starting and ending points.

while loop
while loop goes on till the condition to run the loop is satisfied

Continue
Useful to skip certain iterations of the loop

Break
operator to break a loop

range ()
Creates a range of numbers...
range(start,stop,step)
All three parameters are optional. Start included and stop excluded.

```python
# For loop
icecream = [12,42,54,23]
for jelly in icecream:
    print(jelly+jelly)
```

```python
primes = [2, 3, 5, 7]
for prime in primes:
    print(prime)
```

```python
# Show only the content of the iterator element instead of a list.
for i in range(4):
    print(i)
```

```python
# range() is an iterator you can't print the list directly
print(range(4))
```

```python
# Prints out 3,4,5
for x in range(3, 6):
    print(x)
```

```python
# Prints out 3,5,7
for x in range(3, 8, 2):
    print(x)
```

```python
# while loop
i = 1
while i < 5:
    print('i is: {}'.format(i))
    i = i+1
```

```python
# while loop to print out 0,1,2,3,4
count = 0
while count < 5:
    print(count)
    count += 1 # This is the same as count = count + 1
```

```python
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break
```

```python
# continue skips even numbers prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)
```

```python
# Special else case with while and for loops...
# Prints out 0,1,2,3,4 and then it prints "count value reached 5"
count=0
while(count<5):
    print(count)
    count +=1
else:
    print("count value reached %d" %(count))
```

```python
# Prints out 1,2,3,4
for i in range(1, 10):
    if(i%5==0):
        break
    print(i)
else:
    print("this is not printed because for loop is terminated \
because of break but not due to fail in condition")
```
